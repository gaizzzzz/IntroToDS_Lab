## Exercises 1

Use `re.search` to find whether a string contains a phone number. The pattern that you write should detect a phone number in the following strings.  
```
"Call me at 382-384-3840."  
"my number is (510) 849-3519. Call me!"
```  
And not find a match in the following strings. 
```
"my number is 510-849-35192"  
"here’s my number: 510-849.3519"
``` 
Consider making your own tests as well  

````python
text_1 = '''Call me at 382-384-3840.
my number is (510) 849-3519. Call me!
my number is 510-849-35192
here’s my number: 510-849.3519'''.strip()

for text in text_1.split('\n'):
    regex = r'(\d{3}-\d{3}-\d{4}\b|[(]\d{3}[)]\s*\d{3}-\d{4})'
    match = re.search(regex, text)
    if match is not None:
        print(match.group(0))
````

## Exercise 2

Use `re.sub` to alter the string below so that the dates have a common format that uses a dash for the day, month, and year separator.  
```
03/12/2018, 03.13.18, 03/14/2018, 03:15:2018
```

````python
text_2 = '03/12/2018, 03.13.18, 03/14/2018, 03:15:2018'
dates = text_2.split(', ')
for date in dates:
    print(re.sub(r'[\/\.:]', '/', date))
````

## Exercise 4

In this exercise, you you working with ```email_test.txt``` file (attached), using Regular Expression. 
1. Extract name and email address of sender  
Output should be something like this
```
Name:
"MR. JAMES NGOLA." 
"Mr. Ben Suleman" 
"PRINCE OBONG ELEME" 
"PRINCE OBONG ELEME"

```
```
Email address:
"james_ngola2002@maktoob.com"
"bensul2004nng@spinfinder.com"

"obong_715@epatra.com"
"m_abacha03@www.com"
```
2. Find SUBJECT of All emails from 01/06/2003 to 13/05/2004 


````python
import re
fh = open(r"Data/email_test.txt", "r").read()

# YOUR CODE HERE
# 1. Extract name and email address of sender
emails = fh.strip().split('From r  ')[1:]

address_pattern = r'From: (.+?)\n'
address_info = re.findall(address_pattern, fh)
email_addresses = []
email_pattern = r'\b\w+@[\w\W_]+?\.[\w\W]{2,}\b'
for text in address_info:
    match = re.search(email_pattern, text)
    if match is not None:
        email_addresses.append(match.group(0))
    else:
        email_addresses.append('None')
names = []
for text in address_info:
    name = text.replace(email_addresses[address_info.index(text)], '')
    if '<>' in name:
        name = name.replace('<>', '')
    if '"' in name:
        name = name.replace('"', '')
    name = name.strip()
    names.append(name)
# 2. Find SUBJECT of All emails from 01/06/2003 to 13/05/2004
from datetime import date 
date_subject_pattern = r'Date: (.+?)\nSubject: (.+?)\n|Subject: (.+?)\nDate: (.+?)\n'
result = re.findall(date_subject_pattern, fh)
dates = []
subjects = []
for x in result:
    if x[0] == '':
        subjects.append(x[2])
        dates.append(x[3])
    elif x[3] == '':
        dates.append(x[0])
        subjects.append(x[1])
month_symbols = {'Jan' : 1, 'Feb' : 2, 'Mar' : 3, 'Apr' : 4, 'May' : 5, 'Jun' : 6, 
                 'Jul' : 7, 'Aug' : 8, 'Sep' : 9, 'Oct' : 10, 'Nov' : 11, 'Dec' : 12}
new_dates = []
for d in dates:
    d = ' '.join(d.strip().split())
    date_elements = d.split(' ')
    new_date = date_elements[1:4]
    new_date[1] = month_symbols[new_date[1]]
    if len(new_date[2]) == 2:
        new_date[2] = f'20{new_date[2]}'
    new_date = date(int(new_date[2]), new_date[1], int(new_date[0]))
    new_dates.append(new_date)
dates = new_dates

print('Subject of all emails from 01/06/2003 to 13/05/2004 : ')
for d in dates:
    if date(2003, 6, 1) <= d <= date(2004, 5, 12):
        print(f'Date: {d}, Subject: {subjects[dates.index(d)]}')
````
